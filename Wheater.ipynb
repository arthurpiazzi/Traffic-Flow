```python
import os

dir_path = 'csv/weather'

lista = ['{}/{}'.format(dir_path, item) for item in os.listdir(dir_path) if 'C5988' in item]
lista
```

```python
import pandas as pd
from tqdm import tqdm


    
```

```python
import numpy as np
result = pd.DataFrame()
for f in tqdm(lista):
    print(f)
    df = pd.read_csv(f, infer_datetime_format=True, usecols=['Date_Time', 'air_temp_set_1',
       'relative_humidity_set_1', 'wind_speed_set_1',
       'wind_gust_set_1', 'precip_accum_since_local_midnight_set_1',
       'solar_radiation_set_1', 'precip_accum_24_hour_set_1',
       'dew_point_temperature_set_1d'], index_col = ['Date_Time'])

    df.index =  df.index.str.split(' U').str[0]
    df.index = pd.to_datetime(df.index)
    idx = pd.date_range(start=df.index.min(), end=df.index.max(), freq='5T')
    df = df.reindex(idx, fill_value=np.nan)
    df = df.resample('15T').mean()
    result = result.append(df)

df
```

```python
res = result.interpolate()
result.isna().sum()
```

```python
from pandas.tseries.offsets import DateOffset

def to_sequence_df(df, window):
    df_res = df
    for i in range(1,window):
        df_res = pd.concat([df_res, df.shift(-i)], axis=1, join_axes=[df.shift(-i).index])
        
    df_res.index = df_res.index + DateOffset(hours=4, minutes=45 )
    return df_res[:(len(df_res) -window+1)]
        
r2 = to_sequence_df(res, 20)
r2
```

```python
from sklearn import preprocessing
import numpy as np

def normalize(df1, df2):
    x = np.concatenate((df1.values, df2.values), axis = 0)
    print(x.shape)
    scaler = preprocessing.MinMaxScaler()
    scaler.fit(x)
    return pd.DataFrame(scaler.transform(df1.values)), pd.DataFrame(scaler.transform(df2.values)) 
    
    
x_train , x_test = normalize(x_train, x_test)
y_train, y_test = normalize(y_train, y_test)
```
