```python
import os

dir_path = 'csv/wheater'

lista = ['{}/{}'.format(dir_path, item) for item in os.listdir(dir_path) if 'C5988' in item]
lista
```

```python
import pandas as pd
from tqdm import tqdm


    
```

```python

result = pd.DataFrame()
for f in tqdm(lista):
    print(f)
    df = pd.read_csv(f, infer_datetime_format=True, usecols=['Date_Time', 'air_temp_set_1',
       'relative_humidity_set_1', 'wind_speed_set_1',
       'wind_gust_set_1', 'precip_accum_since_local_midnight_set_1',
       'solar_radiation_set_1', 'precip_accum_24_hour_set_1',
       'dew_point_temperature_set_1d'])
    start, periods = df.Date_Time.min(), len(df)
    df['Date_Time'] = pd.date_range(start=start, freq='5Min', periods=periods)
    df.set_index('Date_Time', inplace=True)
    df = df.resample('15T').mean()
    result = result.append(df)
    

df
```

```python
result.describe(include='all')
```

```python
from pandas.tseries.offsets import DateOffset

def to_sequence_df(df, window):
    df_res = df
    for i in range(1,window):
        df_res = pd.concat([df_res, df.shift(-i)], axis=1, join_axes=[df.shift(-i).index])
        
    df_res.index = df_res.index + DateOffset(hours=4, minutes=45 )
    return df_res[:(len(df_res) -window+1)]
        
r2 = to_sequence_df(result, 20)
r2
```

```python

r3 = result.copy()
r3
```

```python
x_train = r2[r2.index < '2018-02-15 00:00:00']
x_test = r2[((r2.index >= '2018-02-15 00:00:00') & (r2.index < '2018-03-29 23:45:00'))]


y_train = r3[((r3.index >= '2017-10-1 05:00:00') & (r3.index <= '2018-02-15 00:00:00'))]
y_test = r3[((r3.index > '2018-02-15 00:00:00') & (r3.index < '2018-03-30 00:00:00'))]

print(x_train.index.min(), x_train.index.max(), x_train.shape)
print(y_train.index.min(), y_train.index.max(), y_train.shape)
print(x_test.index.min(), x_test.index.max(), x_test.shape)
print(y_test.index.min(), y_test.index.max(), y_test.shape)
```
