```python
import tensorflow as tf
import numpy as np


class RBM(object):
    def __init__(self, n_units = 100):

        self.n_units = n_units
    
    def fit(self, x_train, epochs = 100, batch_size = 100):
        
        n_units = self.n_units
        n_features = x_train.shape[1]
        self.n_features = n_features
        
        vb = tf.placeholder("float", [n_features]) #bias
        hb = tf.placeholder("float", [n_units]) # bias
        W = tf.placeholder("float", [n_features] , n_units]) # weight
        X = tf.placeholder("float", [None, n_features])

        _h0 = tf.nn.sigmoid(tf.matmul(X, W) + hb)  # probabilities of the hidden units
        h0 = tf.nn.relu(tf.sign(_h0 - tf.random_uniform(tf.shape(_h0))))  # sample_h_given_X

        _v1 = tf.nn.sigmoid(tf.matmul(h0, tf.transpose(W)) + vb)
        v1 = tf.nn.relu(tf.sign(_v1 - tf.random_uniform(tf.shape(_v1))))  # sample_v_given_h
        h1 = tf.nn.sigmoid(tf.matmul(v1, W) + hb)

        alpha = 1.0
        w_pos_grad = tf.matmul(tf.transpose(X), h0)
        w_neg_grad = tf.matmul(tf.transpose(v1), h1)
        CD = (w_pos_grad - w_neg_grad) / tf.to_float(tf.shape(X)[0])
        update_w = W + alpha * CD
        update_vb = vb + alpha * tf.reduce_mean(X - v1, 0)
        update_hb = hb + alpha * tf.reduce_mean(h0 - h1, 0)

        err = tf.reduce_mean(tf.square(X - v1))

        cur_w = np.zeros([n_features , n_units], np.float32)
        cur_vb = np.zeros([n_features], np.float32)
        cur_hb = np.zeros([n_units], np.float32)
        prv_w = np.zeros([n_features , n_units], np.float32)
        prv_vb = np.zeros([n_features], np.float32)
        prv_hb = np.zeros([n_units], np.float32)
        sess = tf.Session()
        init = tf.global_variables_initializer()
        sess.run(init)

        sess.run(err, feed_dict={X: x_train, W: prv_w, vb: prv_vb, hb: prv_hb})

        # Parameters
        epochs = 100
        batchsize = 100
        weights = []
        errors = []

        for epoch in range(epochs):
            for start, end in zip(range(0, len(x_train), batchsize), range(batchsize, len(x_train), batchsize)):
                batch = x_train[start:end]
                cur_w = sess.run(update_w, feed_dict={X: batch, W: prv_w, vb: prv_vb, hb: prv_hb})
                cur_vb = sess.run(update_vb, feed_dict={X: batch, W: prv_w, vb: prv_vb, hb: prv_hb})
                cur_hb = sess.run(update_hb, feed_dict={X: batch, W: prv_w, vb: prv_vb, hb: prv_hb})
                prv_w = cur_w
                prv_vb = cur_vb
                prv_hb = cur_hb
                if start % 10000 == 0:
                    self.W = prv
                    self.hb = prv_hb
                    self.vb = prv_vb
                    errors.append(sess.run(err, feed_dict={X: x_train, W: cur_w, vb: cur_vb, hb: cur_hb}))
                    weights.append(cur_w)
            print('Epoch: %d' % epoch, 'reconstruction error: %f' % errors[-1])
            
        self.W = prv
        self.hb = prv_hb
        self.vb = prv_vb
        sess.close()
        
    def predict(self, x):
        hb = tf.placeholder("float", [self.n_units]) # bias
        W = tf.placeholder("float", [self.n_features] , self.n_units]) # weight
        X = tf.placeholder("float", [None, n_features])
        
        with tf.Session as sess:
            sess.run(tf.global_variables_initializer())
            h = tf.nn.sigmoid(tf.matmul(X, W) + hb)
            out_put = sess.run(h, feed_dict={X: x, W: self.W, hb: self.hb})
            return out_put

    
```

```python
class DBN(object):
    def __init__(self, params, n_iter ):
        self.n_layers = len(params)
        self.params = params
        if len(n_iter) == 1:
            self.n_iter = [n_iter]*self.n_layers
        elif len(n_iter) != self.n_layers:
            raise ValueError("The length of params and n_iter must agree")
        else:
            self.n_iter = n_iter
            
    def fit(self, X):
        self.layer = []
        X_input = X
        for i in range(self.n_layers):
            self.layer.append(RBM(self.params[i]))
            self.layer[i].fit(X_input, self.n_iter[i])
            X_input = self.layer[i].predict(X_input)
            
    def predict(self, X):
        X_input = X
        for i in range(self.n_layers):
            X_input = self.layer[i].predict(X_input)
            
        return X_input
            
            
        
```
