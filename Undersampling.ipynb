```python
import os 

root_path = 'csv/traffic'

files = ['{}/{}'.format(root_path, f) for f in os.listdir('csv/traffic') if len(f) < 16]
```

```python
import pandas as pd

shapes = []
for f in files:
    df = pd.read_csv(f)
    try:
        shapes.append([f, df.shape, df.utc_time_id.min(), df.utc_time_id.max()])
    except:
        shapes.append([f, df.shape, df.iloc[:,0].min(), df.iloc[:,0].max()])
        
print(shapes)
```

```python
for s in shapes:
    print(s)
```

```python
import pandas as pd 
from tqdm import tqdm 

# Put your list here
for i,f in tqdm(enumerate(a_list)):
    df = pd.read_csv(f, index_col='utc_time_id', parse_dates=['utc_time_id'])
    df.sort_index(inplace=True)
    df = df.asfreq('5min')
    df = df.resample('15min').mean()
    df = df.apply(pd.to_numeric, downcast='float')
    df.to_csv(f, index=True)

    
    
```
