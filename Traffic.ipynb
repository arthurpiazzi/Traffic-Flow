```python
import os

def get_date(f):
    name, _ = os.path.splitext(f)
    return name[-10:]
```

```python
import pandas as pd 
from tqdm import tqdm
raw_path = "raw"
csv_path = 'csv/traffic'

fields = ['timestamp', 'station', 'District', 'Freeway', 'Direction', 'Lane_type', 'Station_len','Samples', 'per_observed', 'Total_flow', 'Avg_occupancy', 'avg_speed', ]

files = os.listdir(raw_path)
for f in tqdm(files):
    df = pd.read_csv(open('{}/{}'.format(raw_path, f), 'r'), sep=',', header=None, usecols=list(range(12)), names=fields, infer_datetime_format=True, parse_dates=['timestamp'], dtype={'Lane_type':'category', 'station': 'category', 'District': 'category', 'Freeway': 'category', 'Direction': 'category', })
    df.to_csv('{}/{}.csv'.format(csv_path, get_date(f)))

```

```python
lista = ['{}/{}'.format(csv_path, item) for item in os.listdir(csv_path)]

result = pd.DataFrame()
for l in tqdm(lista):
    df = pd.read_csv(l, index_col='timestamp', infer_datetime_format=True)
    df.sort_index(inplace=True)
    df = df.asfreq('5T')
    df = df.resample('15T').mean()
    result = result.append(df)
    
result.to_csv('{}/traffic.csv'.format(csv_path), index=True)
```
